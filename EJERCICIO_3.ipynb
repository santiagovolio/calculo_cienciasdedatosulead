```python
tiende = 3
acerca = limit(1 / (x + 3)**4 , x, 3)
```

# **Gráfica de límite cuando x tiende a 3**

```python
# Rango de valores de x
x_valores = np.linspace(0, 10, 1000)

# Gráfica
fig, ax = plt.subplots()
ax.grid()
ax.plot(x_valores, f(x_valores), label=r"f(x)=$\frac{1}{(x+3)^4}$")

# Punto del límite en x=3

punto_x = tiende

punto_y = float(acerca)

ax.scatter([punto_x], [punto_y], label="Límite en x=3", color='r')

# Señalar valor exacto

plt.legend()
plt.title(r"Gráfico de $f(x)=\frac{1}{(x+3)^4}$")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

```python
from sympy import symbols, limit, sin, pi

t = symbols('t')
limite2 = limit(sin(t + sin(t)), t, pi)
print("Límite exacto:", limite2)
print("Límite decimal:", float(limite2))
```

```python
# Definir la función g(t)

def g(t):
    return np.sin(t + np.sin(t))

# Rango de valores de t

t_vals = np.linspace(0, 2*np.pi, 1000)

# Gráfico
fig, ax = plt.subplots()
ax.grid()
ax.plot(t_vals, g(t_vals), label=r"$g(t)=\sin(t+\sin t)$")

# Señalar el límite

punto_t = np.pi
punto_y = float(limite2)
ax.scatter([punto_t], [punto_y], label="Límite en t=π", color='r')
plt.legend()
plt.title(r"Gráfico de $g(t)=\sin(t+\sin t)$")
plt.ylabel('g(t)')
plt.xlabel('t')
plt.show()
```
