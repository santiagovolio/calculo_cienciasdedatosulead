```python
#Ejercicio 2 (elaborado con IA, la respuesta mía esta en el PDF)

import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(-4, -2, 100)
y1 = 0.5*(x1+4)       # tramo izquierdo

x2 = np.linspace(-2, 0, 100)
y2 = -0.5*(x2+2) + 1  # tramo con hueco

x3 = np.linspace(0, 2, 100)
y3 = -0.5*x3 + 1      # tramo central

x4 = np.linspace(2, 4, 100)
y4 = -0.75*(x4-2)**2 + 1.5  # tramo derecho

x5 = np.linspace(4, 7, 100)
y5 = 0.2*(x5-4) + 0.5       # tramo final

plt.figure(figsize=(8,5))
plt.plot(x1, y1, 'b')
plt.plot(x2, y2, 'b')
plt.plot(x3, y3, 'b')
plt.plot(x4, y4, 'b')
plt.plot(x5, y5, 'b')

# puntos abiertos (discontinuidades)

plt.scatter([-2, 2], [1, 1], color='white', edgecolors='blue', s=80, zorder=5)

# puntos cerrados

plt.scatter([-4, 0, 2, 4], [0, 1, 0.5, 1], color='blue', s=50, zorder=5)

plt.title("Gráfica de f(x) con discontinuidades")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()


x1 = np.linspace(-3, -2, 100)
y1 = -0.5*(x1+3)**2 + 1

x2 = np.linspace(-2, 0, 100)
y2 = (x2+2)**2 - 2

x3 = np.linspace(0.1, 1, 100)
y3 = np.log(x3+0.1)

x4 = np.linspace(1, 3, 100)
y4 = 0.5*(x4-1)**2

plt.figure(figsize=(8,5))
plt.plot(x1, y1, 'g')
plt.plot(x2, y2, 'g')
plt.plot(x3, y3, 'g')
plt.plot(x4, y4, 'g')

# puntos abiertos
plt.scatter([-2, 1], [y2[0], y4[0]], color='white', edgecolors='green', s=80, zorder=5)
# puntos cerrados
plt.scatter([-3, 0, 3], [y1[0], y2[-1], y4[-1]], color='green', s=50, zorder=5)

plt.title("Gráfica de g(x) con discontinuidades")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
#Ejercicio 4

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt


t = sp.symbols('t')

C_base = t**2

def C(t):
  return t**2

def C_der(t):
  return 2*t

C_derivada = sp.diff(C_base, t)

t_0 = 2

y = C_derivada.subs (t, t_0)

#Gráfica

x_valores = np.linspace (-4, 4, 100)

plt.plot(x_valores,C(x_valores), label = "Grafica de C(t)")
plt.plot(x_valores,C_der(x_valores), label = "derivada")
plt.scatter(t_0, y)
plt.legend()
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
plt.show()
```

```python
#Ejercicio 5

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

#Definir variable y funcion

t = sp.symbols('t')

s = t**2 - 8*t + 18

#Derivada para la velocidad

v = sp.diff(s, t)

v_instantanea = v.subs(t, 4)

#Velocidades promedio

# Change intervalos to a NumPy array
intervalos = np.array([(3,4),(3.5, 4), (4, 5), (4, 4.5)])

v_promedios = [(s.subs(t, b) - s.subs(t, a)) / (b-a) for a, b in intervalos]

#Mostrar resultado

print("Velocidad al instante en t = 4: ", v_instantanea)
for i, vp in enumerate(v_promedios, 1):
  print(f"Velocidades promedio en intervalos {intervalos[i - 1]}: {vp}")
```

```python
# Ejercicio 6

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir la variable simbólica y la función de temperatura
# T(t) = (T_inicial - T_refrigerador) * e^(-k*t) + T_refrigerador
tiempo = sp.symbols('tiempo')
temp_inicial = 25  # Temperatura inicial en Celsius
temp_refrigerador = 3  # Temperatura del refrigerador en Celsius
k = 0.1  # Constante de enfriamiento

Temperatura = (temp_inicial - temp_refrigerador) * sp.exp(-k * tiempo) + temp_refrigerador

# Graficar la función de temperatura
valores_tiempo = np.linspace(0, 15, 100)
# Convertir la función simbólica a una función numérica para graficar
funcion_temp_numerica = sp.lambdify(tiempo, Temperatura, 'numpy')
valores_temperatura = funcion_temp_numerica(valores_tiempo)

plt.figure(figsize=(10, 6))
plt.plot(valores_tiempo, valores_temperatura)
plt.xlabel("Tiempo (horas)")
plt.ylabel("Temperatura (°C)")
plt.title("Curva de Enfriamiento de Gaseosa")
plt.grid(True)
plt.show()

# Calcular la razón de cambio (derivada)
Razon_Cambio_Temp = sp.diff(Temperatura, tiempo)
print("Derivada de la función de temperatura (razón de cambio):")
print(Razon_Cambio_Temp)

# Analizar la razón de cambio en t=0 y t=1
razon_en_0 = Razon_Cambio_Temp.subs(tiempo, 0)
razon_en_1 = Razon_Cambio_Temp.subs(tiempo, 1)

print(f"\nRazón de cambio en tiempo=0 (tiempo inicial): {razon_en_0} °C/hora")
print(f"Razón de cambio en tiempo=1 (después de una hora): {razon_en_1} °C/hora")

# Visualizar la razón de cambio (opcional)
funcion_razon_cambio_numerica = sp.lambdify(tiempo, Razon_Cambio_Temp, 'numpy')
valores_razon_cambio = funcion_razon_cambio_numerica(valores_tiempo)

plt.figure(figsize=(10, 6))
plt.plot(valores_tiempo, valores_razon_cambio)
plt.xlabel("Tiempo (horas)")
plt.ylabel("Razón de Cambio de Temperatura (°C/hora)")
plt.title("Razón de Enfriamiento de Gaseosa a lo Largo del Tiempo")
plt.grid(True)
plt.show()

# Conclusión
print("\nConclusión:")
print("La razón de cambio inicial de la temperatura es mayor en magnitud (más negativa) que la razón de cambio después de una hora. Esto significa que la gaseosa se enfría más rápido al principio que después de haber estado en el refrigerador por una hora.")
```

```python
# Ejercicio 6 (ALTERNATIVA PLOTLY)

import plotly.graph_objects as go
import numpy as np
import sympy as sp

# Definir la función de temperatura (reutilizamos la definición de sympy)
tiempo = sp.symbols('tiempo')
temp_inicial = 25  # Temperatura inicial en Celsius
temp_refrigerador = 3  # Temperatura del refrigerador en Celsius
k = 0.1  # Constante de enfriamiento

Temperatura = (temp_inicial - temp_refrigerador) * sp.exp(-k * tiempo) + temp_refrigerador

# Convertir la función simbólica a una función numérica para graficar
funcion_temp_numerica = sp.lambdify(tiempo, Temperatura, 'numpy')

# Generar valores de tiempo y temperatura
valores_tiempo = np.linspace(0, 15, 100)
valores_temperatura = funcion_temp_numerica(valores_tiempo)

# Crear el gráfico de temperatura con Plotly
fig_temp = go.Figure()

fig_temp.add_trace(go.Scatter(x=valores_tiempo, y=valores_temperatura, mode='lines', name='Temperatura'))

fig_temp.update_layout(
    title='Curva de Enfriamiento de Gaseosa (Plotly)',
    xaxis_title='Tiempo (horas)',
    yaxis_title='Temperatura (°C)',
    hovermode='x unified' # Mejora la interactividad al pasar el ratón
)

fig_temp.show()

# Calcular la razón de cambio (derivada)
Razon_Cambio_Temp = sp.diff(Temperatura, tiempo)

# Convertir la función simbólica de la derivada a una función numérica
funcion_razon_cambio_numerica = sp.lambdify(tiempo, Razon_Cambio_Temp, 'numpy')
valores_razon_cambio = funcion_razon_cambio_numerica(valores_tiempo)

# Crear el gráfico de la razón de cambio con Plotly
fig_rate = go.Figure()

fig_rate.add_trace(go.Scatter(x=valores_tiempo, y=valores_razon_cambio, mode='lines', name='Razón de Cambio'))

fig_rate.update_layout(
    title='Razón de Enfriamiento de Gaseosa a lo Largo del Tiempo (Plotly)',
    xaxis_title='Tiempo (horas)',
    yaxis_title='Razón de Cambio de Temperatura (°C/hora)',
    hovermode='x unified' # Mejora la interactividad al pasar el ratón
)

fig_rate.show()
```

```python
#Ejercicio 9

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

t= sp.Symbol('t')

s= t**3 - 6*t**2 + 9*t

s_der1 = sp.diff(s,t)
s_der2 = sp.diff(s_der1,t)

s_base = sp.lambdify(t, s)
s_der1_base = sp.lambdify(t, s_der1)
s_der2_base = sp.lambdify(t, s_der2)

x_valores = np.linspace(-4, 8, 100)
plt.plot(x_valores, s_base(x_valores), label="Funcion base")
plt.plot(x_valores, s_der1_base(x_valores), label="Primera derivada")
plt.plot(x_valores, s_der2_base(x_valores), label="Segunda derivada")
plt.legend()
plt.title("Derivadas de la funcion base")
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
plt.show()
```
